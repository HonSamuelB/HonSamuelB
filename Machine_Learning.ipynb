 # Splitting Data: Train/Test

It involves splitting the whole dataset into:

1.Training data: The algorithm learns from them the hidden pattern.

2.Testing data: The algorithm can test accuracy if the prediction was predicted correctly using the patterns learned previously.
 • Test_size: How much percentage do you want for the test, 0.2 means 20% for testing and 80% for training.
• Random_state: random state number splits the test and training datasets randomly.

# Underfitting:
The model is too simple and fails to capture the data's patterns.

# Overfitting:
The model is too complex and captures noise in the data, performing well on training data but poorly on new data.

 # Underfitting,
 also called undertraining, occurs when the model does not fit the data enough because it fails to find the hidden patterns in the data.

# Overfitting

Also called overtraining, overfitting simply means that our model fits the dataset too well because it has learned to fit the noise, the needless data and patterns that aren't required to make the model work well with other data.

# Overfitting
occurs when a model begins to "memorize"" training data rather than "learning" to generalize from a trend. An overfit model will perform very good on the training set but will perform poorly on new data sets.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
data = {
    'Name': ['John', 'Jack', 'Angelica', 'Andrew', 'Isac'],
             'Last Name': ['Stott', 'Mcbride', 'Newman', 'Hiness', 'Justice'],
             'Age': [10, 20, 15, 15, 29,],
             'Role': ['AI', 'AI', 'web', 'Mobile', 'Manager'],
             'Home Address': ['California', 'California', 'Florida', 'Florida', 'California'],
             'Salary': [0, 1200, 500, 500, 1400],
             'GrLivArea':[0,114,406,100,555]
}

df = pd.DataFrame(data)

x = df[['Age', 'Role', 'Home Address',]]
y = df[['Salary','GrLivArea']]

preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(), ['Role', 'Home Address'])
    ],
    remainder='passthrough'
)

model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', LinearRegression())
])

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)


model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print("Predictions:", y_pred)
print("Actual:", y_test.values)
```

```python
import numpy as np
from scipy import stats


feature = 'GrLivArea'
target = 'SalePrice'
X = data[[feature]].dropna()
y = data.loc[X.index, target]


z_scores = np.abs(stats.zscore(data[[feature, target]].dropna()))

threshold = 3

filtered_data = data[(z_scores < threshold).all(axis=1)]

X_filtered = filtered_data[[feature]]
y_filtered = filtered_data[target]

# plotting the original data with oytliners
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(X, y, alpha=0.5, label='Original data')
plt.xlabel(feature)
plt.ylabel(target)
plt.title('Original Data without Outliers')
plt.legend()

# plotting the filtered data without outliers
plt.subplot(1, 2, 2)
plt.scatter(X_filtered, y_filtered, alpha=0.5, label='Filtered data', color='green')
plt.xlabel(feature)
plt.ylabel(target)
plt.title('Filtered Data without Outliers')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_filtered, y_filtered)

X_range = pd.DataFrame(np.linspace(X_filtered[feature].min(), X_filtered[feature].max(), 100), columns=[feature])
y_pred = model.predict(X_range)

plt.scatter(X_filtered, y_filtered, alpha=0.5, label='Filtered data')
plt.plot(X_range, y_pred, label='Linear regression', color='red')
plt.xlabel(feature)
plt.ylabel(target)
plt.title('Linear Regression on Filtered Data')
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

degree = 4
poly_model = make_pipeline(PolynomialFeatures(degree), LinearRegression())

poly_model.fit(X_filtered, y_filtered)

y_poly_pred = poly_model.predict(X_range)

# Plotting the filtered data points and the polynomial regression line
plt.scatter(X_filtered, y_filtered, alpha=0.5, label='Data points')
plt.plot(X_range, y_poly_pred, label=f'Polynomial regression (degree={degree})', color='red')
plt.xlabel(feature)
plt.ylabel(target)
plt.title('Polynomial Regression on the Filtered Data')
plt.legend()
plt.show()

# plt. savefig('C:/Users/user/Downloads/polynomial_regression_plot.ng')

plt.show()
```
